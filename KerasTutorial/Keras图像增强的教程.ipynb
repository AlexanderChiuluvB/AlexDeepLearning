```python
import keras
# Plot images
from keras.datasets import mnist
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator
```

```python
(X_train,y_train),(X_test,y_test) = mnist.load_data()

plt.figure(figsize=(8,8))

for i in range(6):
    plt.subplot(330+1+i)
    plt.title(y_train[i])
    plt.axis('off')
    plt.imshow(X_train[i],cmap=plt.get_cmap('gray'))
    
    
plt.show()
    
```

随机旋转

```python
(X_train,y_train),(X_test,y_test) = mnist.load_data()
X_train = X_train.reshape(X_train.shape[0],28,28,1)
X_test = X_test.reshape(X_test.shape[0],28,28,1)


datagen = ImageDataGenerator(rotation_range=80)

#要透过训练数据集图像来生成generator的实例
datagen.fit(X_train)

plt.figure(figsize=(8,8))

for X_batch,y_batch in datagen.flow(X_train,y_train,batch_size=9):
    plt.figure(figsize=(8,8))
    for i in range(6):
        plt.subplot(330+1+i)
        plt.title(y_batch[i])
        plt.axis('off')
        plt.imshow(X_batch[i].reshape(28,28),cmap=plt.get_cmap('gray'))

    plt.show()
    break
```

### 随机偏移

参数：width_shift_range， height_shift_range

偏离图框中心点

```python
(X_train,y_train),(X_test,y_test) = mnist.load_data()
X_train = X_train.reshape(X_train.shape[0],28,28,1)
X_test = X_test.reshape(X_test.shape[0],28,28,1)

shift = 0.4
datagen = ImageDataGenerator(width_shift_range=shift,height_shift_range=shift)

#要透过训练数据集图像来生成generator的实例
datagen.fit(X_train)

plt.figure(figsize=(8,8))

for X_batch,y_batch in datagen.flow(X_train,y_train,batch_size=9):
    plt.figure(figsize=(8,8))
    for i in range(6):
        plt.subplot(330+1+i)
        plt.title(y_batch[i])
        plt.axis('off')
        plt.imshow(X_batch[i].reshape(28,28),cmap=plt.get_cmap('gray'))

    plt.show()
    break
```

### 影像特徵標準化 (Feature Standardization)

featurewise_center和featurewise_std_normalization

```python
(X_train,y_train),(X_test,y_test) = mnist.load_data()
X_train = X_train.reshape(X_train.shape[0],28,28,1)
X_test = X_test.reshape(X_test.shape[0],28,28,1)


datagen = ImageDataGenerator(featurewise_center=True, featurewise_std_normalization=True)

#要透过训练数据集图像来生成generator的实例
datagen.fit(X_train)

plt.figure(figsize=(8,8))

for X_batch,y_batch in datagen.flow(X_train,y_train,batch_size=9):
    plt.figure(figsize=(8,8))
    for i in range(6):
        plt.subplot(330+1+i)
        plt.title(y_batch[i])
        plt.axis('off')
        plt.imshow(X_batch[i].reshape(28,28),cmap=plt.get_cmap('gray'))

    plt.show()
    break
```

```python
### ZCA 白化转换

###减少图像矩阵中的冗余和去除相关性的线性特征

(X_train,y_train),(X_test,y_test) = mnist.load_data()
X_train = X_train.reshape(X_train.shape[0],28,28,1)
X_test = X_test.reshape(X_test.shape[0],28,28,1)


datagen = ImageDataGenerator(zca_whitening=True)

#要透过训练数据集图像来生成generator的实例
datagen.fit(X_train)

plt.figure(figsize=(8,8))

for X_batch,y_batch in datagen.flow(X_train,y_train,batch_size=9):
    plt.figure(figsize=(8,8))
    for i in range(6):
        plt.subplot(330+1+i)
        plt.title(y_batch[i])
        plt.axis('off')
        plt.imshow(X_batch[i].reshape(28,28),cmap=plt.get_cmap('gray'))

    plt.show()
    break
```
